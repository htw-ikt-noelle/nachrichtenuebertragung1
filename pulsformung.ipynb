```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
import ipywidgets as widgets

%config InlineBackend.figure_format = 'svg'

plt.style.use('noelle.mplstyle')
plt.rcParams["figure.figsize"] = (4.0,2.5)

#inline, widget 
%matplotlib inline
```

```python
t, w = sym.symbols(r't \omega', real=True)

x = sym.cos(w * t)
# print(sym.latex(x))
#print(x)

w_ = 2 * sym.pi
x_ = sym.lambdify(t, x.subs(w, w_))

t_ = np.arange(-1, 1, 0.01)
plt.plot(t_, x_(t_))
plt.show()
```

```python
t, Ts, alpha = sym.symbols(r't T_s \alpha', real=True, positive=True)
w = sym.symbols(r'\omega', real=True)

h = (sym.sin(sym.pi*t/Ts)) / (sym.pi*t/Ts) * (sym.cos(alpha*sym.pi*t/Ts)) / (1-(2*alpha*t/Ts)**2)
#H = sym.cos(sym.pi*Ts/2/alpha * (w/2/sym.pi - (1-alpha)/2/Ts))**2

Ts_ = 1
alpha_ = 0.1

t_ = np.arange(-5,5,0.01)
h_ = sym.lambdify(t, h.subs(Ts, Ts_).subs(alpha, alpha_))

plt.plot(t_, h_(t_))
plt.show()

#w_ = np.arange(-5,5,0.01)
#H_ = sym.lambdify(w, H.subs(Ts, Ts_).subs(alpha, alpha_))

#plt.plot(w_, H_(t_))
#plt.show()
```
