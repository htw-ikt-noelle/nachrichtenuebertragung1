```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import helpers
import schemdraw
import schemdraw.elements as elm
from schemdraw import dsp
%config InlineBackend.figure_format = 'svg'
plt.style.use('noelle.mplstyle')


d = helpers.gen_block_diagramm()
hl_color = '#FF5555'
hl_label = ['konver']
hl_elm = []
for el in d.elements:
    if len(el._userlabels) == 1:
        if any(exp in el._userlabels[0].label.lower() for exp in hl_label):
            #print(el._userlabels[0].label)
            hl_elm.append(el)

d += elm.lines.EncircleBox(hl_elm, padx=0.5, pady=0.5).color(hl_color).linewidth(3.0).linestyle('--')
d.draw()
```
